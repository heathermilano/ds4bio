# 1. Introduction
The goal of the research article, "RNA Sequencing of Single Human Islet Cells Reveals Type 2 Diabetes Genes," was to use single-cell RNA sequencing to determine the gene expression levels of human pancreatic cells from non-diabetic and type 2 diabetic donors as well as to discover genes with disturbed expression in type 2 diabetic donors. The study included 43,364 genes, so each sample has an expression level of each gene and each sample has metadata about their age, sex, and if they have type 2 diabetes or not.

# 2. Relate Research Questions, Data, and Figures
## 2.1.1 Data and Research Questions
### **Data Set**
This study is taken from Xin 2016 [full article](https://www.cell.com/cell-metabolism/fulltext/S1550-4131(16)30434-X?_returnURL=https%3A%2F%2Flinkinghub.elsevier.com%2Fretrieve%2Fpii%2FS155041311630434X%3Fshowall%3Dtrue)

In the scientific paper, the authors used single-cell RNA sequencing from type 2 diabetics and non-diabetics to reveal human islet cell-type-specific genes. Through their research, they discovered type 2 diabetes affects the expression of 245 genes. In order to complete their research, they obtained 12 non-diabetic donors and 6 type 2 diabetic donors to get 1600 samples. For each of the 1600 samples, they gathered the following information:
* The expression of 43,364 genes
* The age, sex, race, diabetic (or not), and cell type of the sample.

So the observations are the 1600 samples and the features of each sample are the expression of the 43,364 genes. The metadata file includes more information (additional features) for each sample such as the age, sex, race, and cell type of each sample.

The 1600 observations are sequence read archives (SRAs), which are RNA-sequencing reads from the donor's cells. Each sample (observation) is an RNA-sequence read from a donor and each observation has the gene expression level of each gene (feature)*

### **Research Questions**
The research question of the authors was to determine the transcriptomes of 1,492 human pancreatic cells from type 2 diabetic and non-diabetic donors. They looked at the four major endocrine cell types, α cells, β cells, δ cells, and PP cells. With the 1600 samples collected, they identified cell-type-specific genes and pathways in addition to 245 genes with disturbed expression in type 2 diabetes. Most importantly, of the genes they observed, 92% of the genes had not previously been associated with islet cell function or growth. The researchers were able to discover enriched genes and pathways in islet cells as well as islet cells affected by type 2 diabetes. Using the data they collected throughout their research, they were able to find that human islet cells have similar transcriptome profiles with 20 cell-type-specific genes and type 2 diabetes affected the expression of 245 genes.

Another research question I could ask would use the age of the samples to determine if there is a difference between gene expression as diabetic patients age. As humans age, their bodily functions could decrease in functionality, so by using the age of the samples, one could discover that older donros have different gene expressions from younger donors. The research question in the article only compared non-diabetic donors to diabetic donors, but by observing all diabetic donors and comparing their age with their gene expression levels, a new discovery could be made based on the age of diabetic donors and their gene expression. Therefore, we could ask how the age of the diabetic donors affects their gene expression levels.

## 3.3 Visualizations
### 3.3.1 Gene ENSG00000000003
I want to make a histogram to see the number of samples that have an expression level around 0. From the describe function, I know at leaast 50% of the samples have an expression level below 0.5, but I also know there are higher values because the maximum value was almost 73. A histogram allows me to see how many samples have an expression level around 0 and also how many samples have high expression levels.
```python
import seaborn as sns
df = data.transpose() # transpose the data so the genes/features are columns
# data is all the samples and the x-axis is the gene expression levels for the ENSG00000000003 gene
sns.displot(data=df, x=df['ENSG00000000003'])
```

The x-axis is the expression level and the y-axis is the count of the samples that have an expression level between the x-axis bounds of each bin. The describe function showed that over 50% of the samples have an expression level around 0, and the plot confirms this because more than half of the samples (over 800/1600) have an expression level around 0.

This plot also shows the maximum value, which is somewhat hard to see based on the scale of the plot, but there is a small bar at around 73, which was also shown in the describe function as the maximum. This proves that very high expression levels (anything above 20) can be considered outliers because the majority of samples are below 20.
